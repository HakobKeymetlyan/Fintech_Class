# Demo: Intro to Clustering

```python
# Import the modules
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
# Generate synthetic clusters
X, y = make_blobs(
    centers=3, 
    random_state=1
)
```

```python
# Visualize both classes
plt.scatter(X[:, 0], X[:, 1], c=y)
```
